```
import numpy as np
from skimage import io, color
img = io.imread("kitty.jpg")
gray = color.rgb2gray(img)
gray255 = np.multiply(gray, 255)
gray_div = np.divide(gray255.astype(int), 64)
gray_qt = np.multiply(gray_div.astype(int), 64)
gray_div2 = np.divide(gray255.astype(int), 128)
gray_qt2 = np.multiply(gray_div.astype(int), 128)

print(gray255[50,50], gray_div[50,50])


import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()

plt.imshow(gray_qt, cmap="gray")
plt.show()

plt.imshow(gray_qt2, cmap="gray")
plt.show()
```
