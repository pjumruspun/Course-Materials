```
import cv2
import numpy as np
import matplotlib.pyplot as plt


img2 = cv2.imread('week3_img.jpg')
img = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

eq = cv2.equalizeHist(img)

plt.hist(img.flatten(), 256, [0, 256], color='r')
plt.hist(eq.flatten(), 256, [0, 256], color='b')
plt.show()

plt.imshow(img, cmap='gray')
plt.show()

plt.imshow(eq, cmap='gray')
plt.show()
```

```
hist, bins = np.histogram(img.flatten(), 256, [0, 256])
plt.hist(img.flatten(), 256, [0, 256], color='r')

cdf = hist.cumsum()
cdf_norm =  cdf * hist.max() / cdf.max()
plt.plot(cdf_norm, color='b')
plt.show()
```

```
img = cv2.imread('space.png', 0)
eq = cv2.equalizeHist(img)

plt.hist(img.flatten(), 256, [0, 256], color='r')
plt.hist(eq.flatten(), 256, [0, 256], color='b')
plt.show()

plt.imshow(img, cmap='gray')
plt.show()

plt.imshow(eq, cmap='gray')
plt.show()
```

```
kernel = np.ones((3,3), np.float32)/9
dst = cv2.filter2D(img, -1, kernel)

plt.imshow(img, cmap='gray')
plt.show()

plt.imshow(dst, cmap='gray')
plt.show()
```

```
laplacian = np.array([[1,2,1],[2,4,2],[1,2,1]], dtype=np.float32)/16
applied = cv2.filter2D(img, -1, laplacian)

plt.imshow(img, cmap='gray')
plt.show()

plt.imshow(applied, cmap='gray')
plt.show()
```
