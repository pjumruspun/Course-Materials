```
from skimage import io, color
import matplotlib.pyplot as plt
img = io.imread("kitty.jpg")
gray = color.rgb2gray(img)

fig = plt.figure(figsize=(8, 4))
fig.add_subplot(1,2,1)
plt.imshow(img)
fig.add_subplot(1,2,2)
# plt.imshow(gray) # this will show green-ish
plt.imshow(gray, cmap="gray")
io.show()

print(img.shape)
print(gray.shape)

print(f"type(img) = {type(img)}") # type of the object
print(f"img.dtype = {img.dtype}") # type of elements in the object?

print(f"position [50, 50, 2] = {img[50, 50, 2]}")
print(f"position [50, 50] = {img[50, 50]}")
```

```
import cv2
image2 = cv2.imread("kitty.jpg")

print("res1 (weird color):")
plt.imshow(image2)
plt.show()

print(f"position [50, 50] (BGR representation) = {image2[50, 50]}")

cv2.imwrite("weirdtom.jpg", image2)

image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

print("res2 (normal color):")
plt.imshow(image2)
plt.show()

print(f"position [50, 50] (RGB representation) = {image2[50, 50]}")

img2 = cv2.imread("kitty.jpg")
print(f"red = {img2[:,:,2]}")
```

```
# Create Black Image

import numpy as np
import matplotlib.pyplot as plt
import cv2

a = np.zeros((150,150, 3), dtype=int)
# a = 255 + np.ones((150, 150, 3), dtype=int) # White image variation
plt.imshow(a, cmap="gray") 
plt.show()

# Create Red Image

b = np.zeros((100,100, 3), dtype=int)
b[:,:,0] = 255
plt.imshow(b, cmap="gray")
plt.show()

b = cv2.cvtColor(b, cv2.COLOR_BGR_)
cv2.imwrite("red.jpg", b)


```
