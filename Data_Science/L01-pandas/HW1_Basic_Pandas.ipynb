## How many rows are there in the GBvideos.csv after removing duplications?

```python
df = pd.read_csv('data/GBvideos.csv')
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(f'Total rows in GBvideos.csv after removing dupes: {df.shape[0]}')
```

## How many VDO that have "dislikes" more than "likes"?  Make sure that you count only unique title!

```python
dislikes_more_than_likes = df[(df.dislikes > df.likes)][['title', 'dislikes', 'likes']]
dislikes_more_than_likes = dislikes_more_than_likes.drop_duplicates(subset='title')
dislikes_more_than_likes
```

```python
print(f'Number of videos that has more dislike than likes: {dislikes_more_than_likes.shape[0]}')
```

## How many VDO that are trending on 22 Jan 2018 with comments more than 10,000 comments?

```python
df['trending_date'] = pd.to_datetime(df.trending_date, format='%y.%d.%m', errors='ignore', utc=True)
```

```python
from datetime import datetime, timezone

Jan_22_2018 = datetime(2018, 1, 22, tzinfo=timezone.utc)
ans3 = df[(df.trending_date == Jan_22_2018) & (df.comment_count > 10000)]
ans3
```

```python
print(f'Number of videos that are trending on Jan 2018 with comments more than 10,000 comments: {ans3.shape[0]}')
```

## Which date that has the minimum average number of comments per VDO?

```python
df_groupby_date = df.groupby('trending_date')
```

```python
print('The date which has the minimum average number of comments per video is')
df_groupby_date.comment_count.mean().idxmin()
```

## Compare "Sports" and "Comedy", how many days that there are more total daily views of VDO in "Sports" category than in "Comedy" category?

```python
import json
with open('data/GB_category_id.json') as fd:
  cat = json.load(fd)

cat_list = []
for d in cat['items']:
  cat_list.append((int(d['id']), d['snippet']['title']))
```

```python
cat_df = pd.DataFrame(cat_list, columns=['id', 'category'])
```

```python
df_cat = df.merge(cat_df, left_on='category_id', right_on='id')
df_cat[['title', 'category_id', 'category']]
```

```python
groupby_date_cat = df_cat.groupby(['trending_date', 'category'])
```

```python
group_sum = groupby_date_cat.views.sum().reset_index()
```

```python
group_sum
```

```python
sports = group_sum[group_sum.category == 'Sports'][['trending_date', 'views']]
sports.rename(columns={'views': 'sport_views'}, inplace=True)
sports
```

```python
comedy = group_sum[group_sum.category == 'Comedy'][['trending_date', 'views']]
comedy.rename(columns={'views': 'comedy_views'}, inplace=True)
comedy
```

```python
merged = sports.merge(comedy)
print(f'Number of days that daily views of VDO in Sports category is more than in Comedy category: {merged[merged.sport_views > merged.comedy_views].shape[0]}')
```
